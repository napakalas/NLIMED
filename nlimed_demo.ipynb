## The following codes are a demo for NLIMED implementation

```python
from NLIMED import NLIMED
import pprint
query1 = "flux of sodium in basolateral plasma membrane"
query2 = "find all things in repositoy that related to the version of hexokinase activity and ATP as source participant "
```

### <font color=red>Utilising Benepar</font>

```python
nliPmrBenepar = NLIMED(repo='pmr', parser='benepar', pl=2, alpha=3, beta=2, gamma=0.1, delta=0.1, theta=0.3, cutoff=1, tfMode=3)
nliBMOmexBenepar = NLIMED(repo='bm-omex', parser='benepar', pl=2, alpha=3, beta=2, gamma=0.1, delta=0.1, theta=0.3, cutoff=1, tfMode=3)
```

#### Query annotation

```python
pprint.pprint(nliPmrBenepar.getAnnotated(query1))
```

```python
pprint.pprint(nliBMOmexBenepar.getAnnotated(query2))
```

#### NLI to SPARQL

```python
pprint.pprint(nliPmrBenepar.getSparql(query1))
```

```python
pprint.pprint(nliBMOmexBenepar.getSparql(query2))
```
